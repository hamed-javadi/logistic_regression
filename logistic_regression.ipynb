```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#import pdb

url = 'https://raw.githubusercontent.com/hamed-javadi/logistic_regression/master/marks.txt'
data = pd.read_csv(url,heading=None)

h = lambda z: 1 / (1 + np.exp(-z))

def sgd(x,y):
    lr = .001 #learning rate
    x = np.hstack((np.ones((x.shape[0],1)),x))
    epoch = 10000
    m = x.shape[0]
    n = x.shape[1]
    theta = np.random.normal(0,.001,n)
    k = 1
    er = 1
    while k < epoch:
        k = k + 1
        for i in range(m):
            temp = theta
            theta = theta + lr * (y[i] - h(-theta @ x[i])) * x[i]
    return theta

N_train = 200

x_train = 20 * np.random.rand(N_train,1)
y_train = .75 * x_train + 1 + np.random.normal(0,2,(N_train,1))

#pdb.set_trace()
theta = sgd(x_train,y_train)
print('theta[0]=',theta[0],'\ntheta[1]=',theta[1])

x = np.arange(0,x_train.max())
y = theta[0] + theta[1] * x
plt.scatter(x_train,y_train,marker='*')
plt.plot(x,y,'r')
plt.show()
```
